```python
y=df1[['status']]
y
```

```python
X=df1.drop(['status','salary','sl_no'],axis=1)
X[['gender','ssc_b','hsc_b','hsc_s','degree_t','workex','specialisation']] = X[['gender','ssc_b','hsc_b','hsc_s','degree_t','workex','specialisation']].astype('category')
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
ohe.fit_transform(X[['gender','ssc_b','hsc_b','hsc_s','degree_t','workex','specialisation']]).toarray()

```

```python
ohe_array = ohe.fit_transform(X[['gender','ssc_b','hsc_b','hsc_s','degree_t','workex','specialisation']]).toarray()
```

```python
ohe_labels = ohe.categories_
ohe_labels
```

```python
dt = np.array(ohe_labels,dtype='object')
col = np.concatenate(dt)
```

```python
new_data = pd.DataFrame(ohe_array,columns=col)
```

```python
new_data
```

```python
dd= pd.concat([X,new_data],axis=1)
new_dt = dd.drop(['gender','ssc_b','hsc_b','hsc_s','degree_t','workex','specialisation'],axis=1)
```

```python
new_dt
```

```python
y['status'].mask(y['status'] == 'Placed', 1, inplace=True)
```

```python
y['status'].mask(y['status'] == 'Not Placed', 0, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(new_dt, y,train_size=0.8,random_state=1)
print("Input Training:",X_train.shape)
print("Input Test:",X_test.shape)
print("Output Training:",y_train.shape)
print("Output Test:",y_test.shape)
```

```python
X_test[['ssc_p','hsc_p','degree_p',	'etest_p','mba_p']] = (X_test[['ssc_p','hsc_p','degree_p',	'etest_p','mba_p']]/10).apply(np.floor)
X_train[['ssc_p','hsc_p','degree_p',	'etest_p','mba_p']] = (X_train[['ssc_p','hsc_p','degree_p',	'etest_p','mba_p']]/10).apply(np.floor)
```

```python
X_test1 = X_test.astype(int)
X_train1 = X_train.astype(int)
```

```python
y_train1 = y_train.squeeze().astype(int)
y_test1 = y_test.squeeze().astype(int)

```

```python
X_train1
```

## Model Building

###Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=10000)
lr.fit(X_train1, y_train1)
```

```python
y_train_pred = lr.predict(X_train)
y_test_pred = lr.predict(X_test)
lr.score(X_test1, y_test1)
```

###Decision Tree Classifier

```python
from sklearn import tree
dtc = tree.DecisionTreeClassifier()
dtc.fit(X_train1, y_train1)
```

```python
dtc.score(X_test1, y_test1)
```

###Support Vector Machine

```python
from sklearn.svm import SVC
sv = SVC()
sv.fit(X_train1, y_train1)
```

```python
sv.score(X_test1, y_test1)
```

###Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train1, y_train1)
```

```python
rfc.score(X_test1, y_test1)
```

###Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()
mnb.fit(X_train1, y_train1)
```

```python
mnb.score(X_test1, y_test1)
```
